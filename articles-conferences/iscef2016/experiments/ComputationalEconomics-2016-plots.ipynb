```python
path = '../data_csv/data_ticks/'
days = ['august_11','august_12','august_13','august_14','august_15']
assets = ['EURUSD','GBPUSD','USDCHF','USDJPY']
freq = '10s'
ask_list = []
bid_list = []

for day in days:
    data_list = [path + day + '/' + i + '.csv' for i in assets]
    reader = cl.Reader(data_list)
    ticks_list = reader.load()
    ask, bid = reader.resample_list(ticks_list, freq) #,  Para reciproco[2,3]
    ask_list.append(ask)
    bid_list.append(bid)

asks = pd.concat(ask_list)
bids = pd.concat(bid_list)
asks = asks.fillna(method='pad')
bids = bids.fillna(method='pad')
#Common variables
#Y = asks
#N,nassets = Y.shape
#m = cl.Matrix()
#len(asks)
```

```python
def ovecm_it(Y,results,of,L,p,it):
    avg_error = 0; r = 0; N = 1
    n = len(L)
    for j in range(n):
        y = Y[of-L[j]:of+it]
        ovecm = alg.OVECM(y, L[j], p, it, avg_error, r, 1)
        stats = cl.stats(ovecm.y, ovecm.yhat)
        results[j]['MSE'] = np.mean(stats.rmse)**2
        results[j]['MAPE'] = np.mean(stats.mape)
    return results
```

```python
def coint_vect_it(Y,of,L,p,it,freq):
    N,l = Y.shape
    n = len(L)
    m = cl.Matrix()
    results = [{} for k in range(n)]
    rcParams['figure.figsize'] = (18.0, 55.0)
    #fig = plt.figure()
    for j in range(n):
        #fig.add_subplot(n,3,j+1)
        ncoint = []
        for i in range(it):
            y = Y[of+i-L[j]:of+i]
            jres = m.get_johansen(y, p, r=1)
            ncoint.append(jres.ncoint)
        #print ncoint
        count_coint = sum([1 for x in ncoint if(x > 0 and x <4)]) #Count if there are coint
        percentage = count_coint/float(it)*100  # Percentage of cointegration
        results[j]= {'L':L[j],'p':p,'coint':percentage}
        #plt.bar(range(l+1),hist,width=0.5,color='r')
        #plt.xlim([0,l+1])
        #plt.xticks(range(l+2))
        #plt.ylim([0,it])
        #plt.title('Cointegration vectors with L='+str(L[j]))
    #plt.savefig('img/histCointVectors-offset'+str(of)+'-p-'+str(p)+'-freq-'+
    #freq+'.png', format='png', bbox_inches='tight',dpi=100)
    return results
```

```python
def coint_vect_plot(Y,of,L,p,it,freq):
    matplotlib.rcParams.update({'font.size': 15})
    N,l = Y.shape
    n = len(L)
    m = cl.Matrix()
    results = [{} for k in range(n)]
    fig, axs = plt.subplots(2,2,figsize=(13, 9),facecolor="white")
    axs = axs.ravel()
    width = 0.6
    for j in range(n):
        ncoint = []
        for i in range(it):
            y = Y[of+i-L[j]:of+i]
            jres = m.get_johansen(y, p, r=1)
            ncoint.append(jres.ncoint)
        #print ncoint
        hist, axis = np.histogram(ncoint,bins=np.arange(l+2),density=True)
        print hist
        axs[j].bar(np.arange(l+1)-width/2.0,hist,width,color='k')
        axs[j].set_xlim([-width,l+width])
        axs[j].set_ylim([0,1.0])
        axs[j].set_title('Cointegration vectors with L='+str(L[j]))
        
        count_coint = sum([1 for x in ncoint if(x > 0 and x <4)]) #Count if there are coint
        percentage = count_coint/float(it)*100  # Percentage of cointegration
        results[j]= {'L':L[j],'p':p,'coint':percentage}
    plt.savefig('img/histCointVectors-offset'+str(of)+'-p-'+str(p)+'-freq-'+
                freq+'.eps', format='eps', bbox_inches='tight',dpi=1200)
    return results
```
