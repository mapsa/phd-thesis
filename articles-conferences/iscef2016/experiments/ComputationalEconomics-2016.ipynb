```python
def ECM(y,P):
    """
    Classical VECM
    """
    m = cl.Matrix()
    y_i = y[:-1]
    
    beta = m.get_johansen(y_i.as_matrix(), P, 1)
    A, B = m.vec_matrix(y_i, P, beta.evecr)

    x, residuals, rank, s = np.linalg.lstsq(A, B)
    A, B = m.vec_matrix(y, P, beta.evecr)
    
    yhat = np.dot(A[-1,:], x) + y.as_matrix()[-2:-1]
    ytrue = y[-1:].as_matrix()

    return yhat,ytrue
```

```python
path = '../data_csv/data_ticks/'
days = ['august_11','august_12','august_13','august_14','august_15']
assets = ['EURUSD','GBPUSD','USDCHF','USDJPY']
freq = '10s'
ask_list = []
bid_list = []

for day in days:
    data_list = [path + day + '/' + i + '.csv' for i in assets]
    reader = cl.Reader(data_list)
    ticks_list = reader.load()
    ask, bid = reader.resample_list(ticks_list, freq) #,  Para reciproco[2,3]
    ask_list.append(ask)
    bid_list.append(bid)

asks = pd.concat(ask_list)
bids = pd.concat(bid_list)
asks = asks.fillna(method='pad')
bids = bids.fillna(method='pad')
#Common variables
Y = asks
N,nassets = Y.shape
m = cl.Matrix()
len(asks)
```

# AVECM

```python
def format_results(stats):
    new_line=''
    STATS = [stats.mape,stats.mae,stats.rmse]
    for y in STATS:
        for x in y:
            new_line=new_line+",{:.3e}".format(x)
        new_line = new_line+",{:.3e}".format(np.mean(y))
    return new_line
```

```python
def get_best_params(pms,it,Y,of,method):
    n = len(pms)
    m = cl.Matrix()
    results = []
    for j in range(n):
        l, p = pms[j]
        if method == 'coint':
            ncoint = []
            for i in range(it):
                y = Y[of-i-l:of-i]
                try:
                    jres = m.get_johansen(y, p, r=0)  # ACA SE CAE
                    ncoint.append(jres.ncoint)
                except:
                    print of,j,pms[j],i
                    pass 
            count_coint = sum([1 for x in ncoint if(x > 0 and x <4)]) #Count if there are coint
            percentage = count_coint/float(it)*100  # Percentage of cointegration
            results.append({'l':l,'p':p,'coint':percentage})
            if percentage == 100.0:
                break
        elif method == 'MAPE':
            y = Y[of-it-l:of]
            ovecm = alg.OVECM(y, l, p, it, 0, 0, 1)
            stats = cl.stats(ovecm.y, ovecm.yhat)
            mape = mean(stats.mape)
            results.append({'l':l,'p':p,'MAPE':mape})
    best_params = max(results, key=lambda x:x[method])
    return best_params['l'],best_params['p'],best_params[method]
```

```python
nday = 1; hour = 360; day = 8640
of = nday*day  + 9*hour
it = 10       # In-sample iterations
outit = 20    # Out of sample iterations
nparams = 47  # Number of possible combinations of L and P
method = 'coint'  # 'MAPE'
#method = 'MAPE'  # 'MAPE'
L = range(100,4000,4000/(nparams/5))
P = range(1,4)
pms = [[l, p] for p in P for l in L ]

Ytrue = np.zeros([outit, nassets], dtype=np.float32)
Yhat = np.zeros([outit, nassets], dtype=np.float32)

start_time = time.time()
coint_hist = np.zeros(outit, dtype=np.int)
for i in range(outit):
    l, p, coint = get_best_params(pms,it,Y,of+i,method)
    coint_hist[i] = coint
    y = Y[of+i+1-l:of+i+1]
    Yhat[i,:],Ytrue[i,:] = ECM(y,p)

tot_time = time.time() - start_time
print tot_time
stats = cl.stats(Ytrue[1:], Yhat[1:])
print "Offset,it,nparams,MAPE,MAE,RMSE,TIME"
print of,it,nparams,format_results(stats),tot_time/outit
```

## Augmented Dickey Fuller test

```python
def ADF(v, crit='5%', max_d=6, reg='nc', autolag='AIC'):
    """ Augmented Dickey Fuller test

    Parameters
    ----------
    v: ndarray matrix
        residuals matrix

    Returns
    -------
    bool: boolean
        true if v pass the test 
    """

    boolean = False

    
    adf = adfuller(v, max_d, reg, autolag)
    print adf
    if(adf[0] < adf[4][crit]):
        pass
    else:
        boolean = True

    return boolean
```
