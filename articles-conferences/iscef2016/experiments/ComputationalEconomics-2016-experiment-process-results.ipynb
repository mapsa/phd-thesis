```python
assets = ['EURUSD','GBPUSD','USDCHF','USDJPY']
pylab.rcParams['figure.figsize'] = (15.0, 10.0)
nday = 4; hour = 360; day = 24*hour
OF = [x+day for x in range(0,nday*day,hour)];OF = OF[:-3]  #93 time intervals
rango = range(12 + 5*hour)  # Number of lines of results files
outit=360
#OF = OF[:8]
```

```python
STATS=[]
stats_best = []
stats_rw = []
stats_ar = []
frame_y = []
frame_avecm = []
frame_ar = []
frame_rw = []
frame_coint = []
columns = ['Offset','Iterations','Nparams',
                  'MAPE-EURUSD','MAPE-GBPUSD','MAPE-USDCHF','MAPE-USDJPY','MAPE-MEAN',
                  'MAE-EURUSD','MAE-GBPUSD','MAE-USDCHF','MAE-USDJPY','MAE-MEAN',
                  'RMSE-EURUSD','RMSE-GBPUSD','RMSE-USDCHF','RMSE-USDJPY','RMSE-MEAN']
for i in range(len(OF)):
    #print OF[i]
    filename = 'vtests/results-offset-'+str(OF[i])+'.txt'
    df = pd.read_csv(filename,skiprows=[x for x in rango if x>8],
                     header=None,names=columns,index_col=['Offset','Iterations','Nparams'])
    STATS.append(df)
    # BEST RUNS
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<9 or x>9],
                          header=None,names=columns, usecols=range(1,19),index_col='Offset')
    #df_best.columns = columns
    stats_best.append(df)
    # RW
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<10 or x>10],
                        header=None,names= ['Offset']+columns[3:],usecols=range(1,17),index_col='Offset')
    stats_rw.append(df)
    # ARIMA
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<11 or x>11],
                        header=None,names= ['Offset']+columns[3:],usecols=range(1,17),index_col='Offset')
    stats_ar.append(df)
    # Y REAL
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<12 or x>11+outit],
                        header=None,names=assets)
    frame_y.append(df)
    # Y AVECM
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<12+outit or x>11+2*outit],
                        header=None,names=assets)
    frame_avecm.append(df)
    # Y AR
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<12+2*outit or x>11+3*outit],
                        header=None,names=assets)
    frame_ar.append(df)
    # Y RW
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<12+3*outit or x>11+4*outit],
                        header=None,names=assets)
    frame_rw.append(df)
    # Y RW
    df = pd.read_csv(filename,skiprows=[x for x in rango if x<12+4*outit or x>11+5*outit],
                        header=None,names=['Percentage'])
    frame_coint.append(df)
results = pd.concat(STATS)
results_best = pd.concat(stats_best)
results_rw = pd.concat(stats_rw)
results_ar = pd.concat(stats_ar)
y = pd.concat(frame_y)
avecm = pd.concat(frame_avecm)
ar = pd.concat(frame_ar)
rw = pd.concat(frame_rw)
coint = pd.concat(frame_coint)
```

# Final Stats all data

```python
Y = y
AVECM = avecm
AR = ar
RW = rw
results = cl.stats(Y,AVECM)
MSE_avecm = results.rmse*results.rmse
print 'MSE AVECM',MSE_avecm
results = cl.stats(Y,AR)
MSE_ar = results.rmse*results.rmse
print 'MSE AR', MSE_ar
results = cl.stats(Y,RW)
MSE_rw = results.rmse*results.rmse
print 'MSE RW',MSE_rw

#Calculates the T-test on TWO RELATED samples of scores, a and b.
#This is a two-sided test for the null hypothesis that 2 related 
#or repeated samples have identical average (expected) values.

e_avecm = Y-AVECM; e2_avecm = e_avecm*e_avecm
e_ar = Y-AR; e2_ar = e_ar*e_ar
ttest = stats.ttest_rel(e2_avecm,e2_ar)
print 'T test MSE',ttest[1]

diff_percentual =(MSE_avecm/MSE_ar-1)*100
print 'diff_percentual', diff_percentual
```

```python
def stats_plot(assets,results_best,results_rw,results_ar,statsname):
    fig, axs = plt.subplots(2,2,figsize=(15, 6))
    axs = axs.ravel()
    for i in range(len(assets)):
        obj_name=statsname+'-'+assets[i]
        axs[i].plot(range(len(OF)),results_best[obj_name]/results_rw[obj_name],'b-')
        axs[i].plot(range(len(OF)),results_ar[obj_name]/results_rw[obj_name],'r-')
        axs[i].set_title(obj_name)
        axs[i].set_ylim([0.4,1])
    return fig
```
